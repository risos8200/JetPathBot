```python
from collections import defaultdict 
graph = defaultdict(list)
for edge in edges:
    a, b = edge[0], edge[1]
    graph[a].append(b)
    graph[b].append(a)

```

```python
def BFS_SP(graph, start, goal):
    explored = []
    queue = [[start]]
    if start == goal:
        print("Same Node")
        return
    while queue:
        path = queue.pop(0)
        node = path[-1]
        if node not in explored:
            neighbours = graph[node]
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)
                if neighbour == goal:
                    return(new_path)
            explored.append(node)
    print("So sorry, but a connecting path doesn't exist :(")
    return
path1=BFS_SP(graph, 5, 20)
path2=BFS_SP(graph, 20, 31)
```
